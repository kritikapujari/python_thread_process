```python
#Process that run in parallel
#CPU Bound Tasks-Tasks that are heavy on CPU usuage
#Parallel execution-Multipke cores of CPU
```

```python
import multiprocessing
import time
```

```python
def square_numbers():
    for i in range(5):
        time.sleep(1.5)
        print(f"Cube:{i * i }")
```

```python
def cube_numbers():
    for i in range(5):
        time.sleep(1.5)
        print(f"Cube:{i * i * i}")
```

```python
 #create 2 processes

if __name__=="__main__":
    p1=multiprocessing.Process(target=square_numbers)
    p2=multiprocessing.Process(target=cube_numbers)
    t = time.time()

    p1.start()
    p2.start()


    p1.join()
    p2.join()

    finished_time=time.time()-t
    print(finished_time)

```
